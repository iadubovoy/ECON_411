The quantity theory of money begins with the equation of exchange:

\begin{equation*}
\mathbf{MV=PY}
\label{eq:equation1} \tag{1}
\end{equation*}
 
where M is the total supply of Money in an economy, V is the velocity of the circulation of money in the economy, P is the price level of all goods and services in the economy, and Y is the real income. Let's dive into what equation (1) is telling us. It is saying that the amount of money in an economy multiplied by the speed by which the money is being used is equal to the price of the goods/services multiplied by the real income. There are implications that we can use equation (1) to help us understand and use other formulas in our understanding of the economy. Furthermore, it might be possible to extend the quantity theory of money to displines outside the realm of economics (Allais, 1966). Since money is crucial to modern economies, understanding how it works with other situations helps us to make better futures. There are many ways to caluclate and imagine the amount of money in an economy, which is why M can be calcualted multiple ways (Friedman, 1970). We will be looking at a couple of the ways that money can be calulated in a little bit. The two variables for money that we will be looking are the monetary base and the M4 variable. The monetary base is the money that is in circulation that is readily available such as in people's hands or owned by the central bank. M4 is a broader money index that includes monetary bas plus less liquid forms of money such as bonds. For now, the main goal is to test the equation the equation of exchange. If equation (1) holds to be true, then algebra should be able to be used to test other forms of equation (1). There are numerous forms of the equation of exchange. However, we will only look at perhaps the most intuitive transformations of equation (1). We are only interested in the equations where a particular variable is multiplied or divided such that variable times/divided 1 will result in a unique equation. The goal of these operations will be to test the equation of exchange in a rigorous format. It seems that if the equation of exchange holds true, then all algebraic changes should also hold true. If there are cases where the new equations do not hold true, then it should be investigated as to why they are false. A possible reason as to why it is importnant to alter the form of this equation is to test that this equation is still still true when certain variables are in the denominator. One mathematical posibility is that one of the variables equals 0 in which case it would not be an equality below but equation (1) would be an equality. Thus, they would not be equal. This analysis desires to be as thorough as possible in testing of the the equation of exchange.

```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt

def gather_data(data_codes, start, end = datetime.datetime.today(), freq = "A"):
    i = 0
    for key, val in data_codes.items():
        if i == 0:
            df = web.DataReader(val, "fred", start, end).resample(freq).mean()
            df.rename(columns = {val:key}, inplace = True) 
            i = None
        else:
            df[key] = web.DataReader(val, "fred", start, end).resample(freq).mean()

    return df

data_codes  = {"Mb": "MBCURRCIR",
               "M4": "MZM",
               "r": "DFF"}

start = datetime.datetime(1992, 1, 1)
end = datetime.datetime(2021, 2, 1)

data = gather_data(data_codes = data_codes, start = start, end = end, freq = "M")
```

```python
data[["(PY)/V"]] = data["PY"].div(data["V"])
data[["M/Y"]] = data["M"].div(data["Y"])
data[["P/V"]] = data["P"].div(data["V"])
data[["M/(PY)"]] = data["M"].div(data["PY"])
data[["1"]] = data["M"].div(data["M"])
data[["1/V"]] = data["1"].div(data["V"])
data[["Y/V"]] = data["Y"].div(data["V"])
data[["M/P"]] = data["M"].div(data["P"])
data[["(MV)/P"]] = data["MV"].div(data["P"])
data[["(MV)/(PY)"]] = data["MV"].div(data["PY"])
data[["(MV)/Y"]] = data["MV"].div(data["Y"])
data[["V/Y"]] = data["V"].div(data["Y"])
data[["P/M"]] = data["P"].div(data["M"])
data[["(PY)/M"]] = data["PY"].div(data["M"])
data[["V/P"]] = data["V"].div(data["P"])
data[["Y/M"]] = data["Y"].div(data["M"])
data[["V/(PY)"]] = data["V"].div(data["PY"])
data[["1/M"]] = data["1"].div(data["M"])
data[["1/(PY)"]] = data["1"].div(data["PY"])
data[["1/(MV)"]] = data["1"].div(data["MV"])
data[["1/P"]] = data["1"].div(data["P"])
data[["Y/(MV)"]] = data["Y"].div(data["MV"])
data[["(PY)/(MV)"]] = data["PY"].div(data["MV"])
data[["1/Y"]] = data["1"].div(data["Y"])
data[["P/(MV)"]] = data["P"].div(data["MV"])
```

Even though all the data is very close to 0, there appears to be different types of fluctation that occurs in the graphs. This does not seem to be significant. But let's run a basic t-test in order to prove where the data is. We will use a library called scipy.stats for our tests.

```python
from scipy.stats import ttest_1samp
```

We will create three differnet functions that run the t-test drom scipy.stats. We are using a one sample t-test which means that our null hypothesis is comparing one set of data to a mean. We will create 3 functions, and each function will have a different value for the mean the sample is being compared to: we have negative one billionth, 0, and one billionth. All very small numbers, and I will demonstrate why I chose those specific values in a little bit. First, let's make the functions.

```python
def ttest1(list_obj):
    for key in list_obj: 
        ttest1_ = ttest_1samp((list_obj), -0.000000001)
    
    return ttest1_

def ttest2(list_obj):
    for key in list_obj: 
        ttest2_ = ttest_1samp((list_obj), 0)
    
    return ttest2_

def ttest3(list_obj):
    for key in list_obj: 
        ttest3_ = ttest_1samp((list_obj), 0.000000001)
    
    return ttest3_
```

Now we will create 3 different tables with every null hypothesis having there own table. In it we will compare the differnce of every equation to the null hypothesis such that there will be 16 rows of data per table.

```python
table1 = pd.DataFrame(ttest1(updated_data))
table1 = table1.T
table1.rename(columns = {0:"T-stat", 1:"P-value"}, inplace = True)
table1.rename(index = {0:'EQ1', 1:'EQ2', 2:'EQ3', 3:'EQ4', 4:'EQ5', 5:'EQ6', 6:'EQ7', 7:'EQ8', 8:'EQ9', 9:'EQ10', 10:'EQ11', 11:'EQ12', 12:'EQ13', 13:'EQ14', 14:'EQ15', 15:'EQ16'}, inplace = True)
table1
```

```python
table2 = pd.DataFrame(ttest2(updated_data))
table2 = table2.T
table2.rename(columns = {0:"T-stat", 1:"P-value"}, inplace = True)
table2.rename(index = {0:'EQ1', 1:'EQ2', 2:'EQ3', 3:'EQ4', 4:'EQ5', 5:'EQ6', 6:'EQ7', 7:'EQ8', 8:'EQ9', 9:'EQ10', 10:'EQ11', 11:'EQ12', 12:'EQ13', 13:'EQ14', 14:'EQ15', 15:'EQ16'}, inplace = True)
table2 
```

```python
table3 = pd.DataFrame(ttest3(updated_data))
table3 = table3.T
table3.rename(columns = {0: "T-stat", 1:"P-value"}, inplace = True)
table3.rename(index = {0:'EQ1', 1:'EQ2', 2:'EQ3', 3:'EQ4', 4:'EQ5', 5:'EQ6', 6:'EQ7', 7:'EQ8', 8:'EQ9', 9:'EQ10', 10:'EQ11', 11:'EQ12', 12:'EQ13', 13:'EQ14', 14:'EQ15', 15:'EQ16'}, inplace = True)
table3
```

Now we wish to demonstrate that the above test where we proved that the equation of exchange is true would still be true if we used M4. The logic behind it is as follows:
\begin{align}
 \begin{aligned}
    M_{b}V_{b} &= PY \\
    M_{4}V_{4} &= PY & \mbox{means that} \\
    M_{b}V_{b} &= PY = M_{4}V_{4} & \mbox{such that} \\
 \end{aligned}
\end{align}
\begin{equation*}
\mathbf{M_bV_b=M_4V_4}
\label{eq:equation19} \tag{19}
\end{equation*} 

We need to create a few variables in order to test if equation (19) is true.

```python
data[["MbVb"]] = data["Mb"] * data["Vb"]
data[["M4V4"]] = data["M4"] * data["V4"]
data[["Base - M4"]] = data["MbVb"] - data["M4V4"]
```

Now we wish to try to find a realtionship between money and interest rate. The reasoning behind this is easier to understand with these equations that Friedman (1971) wrote about. The equations are this:
\begin{equation*}
\mathbf{M_D=M_S}
\label{eq:equation20} \tag{20}
\end{equation*} 

\begin{equation*}
\mathbf{M_S=h(r)}
\label{eq:equation21} \tag{21}
\end{equation*} 

\begin{equation*}
\mathbf{M_D=Y*l(r)}
\label{eq:equation22} \tag{22}
\end{equation*} 
This tells us that money demanded is equal to money supply in equilibrium, this is the ideal place to be in this system. It also tells us that the money supplied is a function of the interest rate, so we could say that money demanded is a function of the interest rate. We also see that money demanded is equal to real icome times a function of the interest rate. This indicates that real income adjusted with interest rate is equal to the money demanded. These are all interesting equations, and let's run some tests to see if we can spot any of these relationships. Statsmodels.api is used to run the regressions.

```python
import statsmodels.api as sm
model = sm.OLS.from_formula('M ~ r', data)
result = model.fit()
result.summary()
```

This shows that there is some corelation between M and r, but the r^2 makes me want to try another method. I will log the data and compare it that way.

```python
data[["lnM"]] = np.log(data["M"])
data[["lnr"]] = np.log(data["r"])
data
```

The newly created variables were sucsefully made, and we will create a plot with them.

```python
data[["lnM", "lnr"]].plot.line()
```

The rate of money being increased is growing over time while the interest rate is realtively stable outside of the finacial crisis and COVID-19. 

```python
model = sm.OLS.from_formula("lnM ~ lnr", data)
result = model.fit()
result.summary()
```

This test was very similar to the one shown earlier, but it has slightly better r squared value. Now, we will try to appl equation (22) to the model. We will log real income first.

```python
data[["lnY"]] = np.log(data["Y"])
```

Now we will see the corelation between the natural log of money and the natural log of real income.

```python
model = sm.OLS.from_formula("lnM ~ lnY", data)
result = model.fit()
result.summary()
```

I will add the natural log of interest rate to the function to see if there is any difference in the result of the regression.

```python
model = sm.OLS.from_formula("lnM ~ lnr + lnY", data)
result = model.fit()
result.summary()
```

The r squared is better which is expected, and the natural log of interest rate is statically significant such that for every 1% increase to the interest rate, the quantity of money decreases by 3.72%. This is an imporant result, and it is consistent with theory and pass results with different datasets (Laidler, 1966).
