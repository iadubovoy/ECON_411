```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt

def gather_data(data_codes, start, end = datetime.datetime.today(), freq = "A"):
    i = 0
    for key, val in data_codes.items():
        if i == 0:
            df = web.DataReader(val, "fred", start, end).resample(freq).mean()
            df.rename(columns = {val:key}, inplace = True) 
            i = None
        else:
            df[key] = web.DataReader(val, "fred", start, end).resample(freq).mean()

    return df

def bil_to_mil(series):
    return series.mul(10**3)

data_codes  = {"Core CPI": "CPILFESL",
               "Monetary Base (Millions of Dollars)": "BOGMBASE"}

start = datetime.datetime(1960, 1, 1)
end = datetime.datetime.today()

data = gather_data(data_codes = data_codes, start = start, end = end, freq = "Y")
```

```python
data[["Monetary Base (Billions of Dollars)"]] = (data["Monetary Base (Millions of Dollars)"]).div(1000)
updated_data = data[["Core CPI", "Monetary Base (Billions of Dollars)"]]
updated_data
```

## What is the difference between Fiscal and Monetary policy?

### Fiscal policy are the policies that deal with taxation and government spending, so government revenue and government expenditures. These policies usually come through the legislative or the exuctive branch. It is noteworthy that federal, state, and local governments could have fiscal policies that vary greatly. Monetary policy deals with the federal reserves making policy that shape their macroeconomic goals. Some gaols could be to decrease the price level or increase inflation.

## An introduction into how fiscal policy is created from a public choice perspective

###  Fiscal policies are made in a game of polictics, and it is rational to assume that people make tradeoffs in politics as political players tend to make agreements in order to try to accomplish their main goals. They will often give on issues that are of not as high importnance to them, such as a large, coastal city congressperson giving in on an agricultural issue and gaining support on a pro-port bill from a midwestern congressperson. These types of exchanges can be thought of as similar to positive-sum exchanges that are seen in everyday life. 

### Given this basic framework in mind, it is easy to see how Fiscal policy issues could work in a way that is counter-productive. This could happen if a Fiscal policy is agreed upon by the majority (e.g., the bill is passed) of the legislation. But in the process, there are going to be people that will vote for other bills that previosly they would not have voted for, and this sort of calling in the favors could lead to two bills being passed that offset the goals of eachh other. Suppose that one bill was passed that increased income tax and another bill was passed that decreased sales tax. This might have kept the overall revenue for the governemnt at about the same, depsite the (assumed) intention of one of the bills was to increase revenue. Likewise the other bill wanted to decrease governmnet income, and it failed too.

### It is possible to see that policial players in the obove scenario could make a positive-sum exchange; therefore, they are motivated to behave in a similar way in the future. Wile the overall outcome might have had some reprecusions in the short-term, such as one bill going into effect before the other bill, but in this example there was a neutral benefit (in terms of the perceived goals of the political players). But since the players are incentivized to continue this trend, it likely will continue.
