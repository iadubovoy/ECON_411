## Equation of Exchange

The quantity theory of money begins with the equation of exchange:

\begin{equation*}
\mathbf{MV=PY}
\label{eq:equation1} \tag{1}
\end{equation*}
 
where M is the total supply of Money in an economy, V is the velocity of the circulation of money in the economy, P is the price level of all goods and services in the economy, and Y is the real income. Let's dive into what equation (1) is telling us. It is saying that the amount of money in an economy multiplied by the speed by which the money is being used is equal to the price of the goods/services multiplied by the real income. There are implications that we can use equation (1) to help us understand and use other formulas in our understanding of the economy. Furthermore, it might be possible to extend the quantity theory of money to displines outside the realm of economics. [6] Since money is crucial to modern economies, understanding how it works with other situations helps us to make better futures. There are many ways to caluclate and imagine the amount of money in an economy, which is why M can be calcualted multiple ways. [1] If equation (1) holds to be true, then algebra should be able to be used to test other forms of equation (1). There are numerous forms of the equation of exchange. However, we will only look at perhaps the most intuitive transformations of equation (1). We are only interested in the equations where a particular variable is multiplied or divided such that variable times/divided 1 will result in a unique equation. The goal of these operations will be to test the equation of exchange in a rigorous format. It seems that if the equation of exchange holds true, then all algebraic changes should also hold true. If there are cases where the new equations do not hold true, then it should be investigated as to why they are false. A possible reason as to why it is importnant to alter the form of this equation is to test that this equation is still still true when certain variables are in the denominator. One mathematical posibility is that one of the variables equals 0 in which case it would not be an equality below but equation (1) would be an equality. Thus, they would not be equal.
```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt

def gather_data(data_codes, start, end = datetime.datetime.today(), freq = "A"):
    i = 0
    for key, val in data_codes.items():
        if i == 0:
            df = web.DataReader(val, "fred", start, end).resample(freq).mean()
            df.rename(columns = {val:key}, inplace = True) 
            i = None
        else:
            df[key] = web.DataReader(val, "fred", start, end).resample(freq).mean()

    return df

def bil_to_mil(series):
    return series.mul(10**3)

data_codes  = {"M": "M1SL",
               "V": "M1V",
               "P": "GDPDEF",    
               "Y": "GDPC1"}

start = datetime.datetime(1950, 1, 1)
end = datetime.datetime.today()

data = gather_data(data_codes = data_codes, start = start, end = end, freq = "Q")
```

But we will move on and try to find some more answers to our question with the data. The next step is to create new variables that will represent each side of equations (1) through equation (16). We will just manipulate Pandas DataFrames and create new varaibles similar to the process described earlier.

```python
data[["(PY)/V"]] = data["PY"].div(data["V"])
data[["M/Y"]] = data["M"].div(data["Y"])
data[["P/V"]] = data["P"].div(data["V"])
data[["M/(PY)"]] = data["M"].div(data["PY"])
data[["1"]] = data["M"].div(data["M"])
data[["1/V"]] = data["1"].div(data["V"])
data[["Y/V"]] = data["Y"].div(data["V"])
data[["M/P"]] = data["M"].div(data["P"])
data[["(MV)/P"]] = data["MV"].div(data["P"])
data[["(MV)/(PY)"]] = data["MV"].div(data["PY"])
data[["(MV)/Y"]] = data["MV"].div(data["Y"])
data[["V/Y"]] = data["V"].div(data["Y"])
data[["P/M"]] = data["P"].div(data["M"])
data[["(PY)/M"]] = data["PY"].div(data["M"])
data[["V/P"]] = data["V"].div(data["P"])
data[["Y/M"]] = data["Y"].div(data["M"])
data[["V/(PY)"]] = data["V"].div(data["PY"])
data[["1/M"]] = data["1"].div(data["M"])
data[["1/(PY)"]] = data["1"].div(data["PY"])
data[["1/(MV)"]] = data["1"].div(data["MV"])
data[["1/P"]] = data["1"].div(data["P"])
data[["Y/(MV)"]] = data["Y"].div(data["MV"])
data[["(PY)/(MV)"]] = data["PY"].div(data["MV"])
data[["1/Y"]] = data["1"].div(data["Y"])
data[["P/(MV)"]] = data["P"].div(data["MV"])
```

We can now create new variables that describe the difference between the left hand side of an equation and the right hand side of that equation. So, we will create 16 new variables, and we will do some analysis with these newly created variables.

```python
data[["EQ1"]] = data["MV"] - data["PY"]
data[["EQ2"]] = data["M"] - data["(PY)/V"]
data[["EQ3"]] = data["M/Y"] - data["P/V"]
data[["EQ4"]] = data["M/(PY)"] - data["1/V"]
data[["EQ5"]] = data["M/P"] - data["Y/V"]
data[["EQ6"]] = data["(MV)/P"] - data["Y"]
data[["EQ7"]] = data["(MV)/(PY)"] - data["1"]
data[["EQ8"]] = data["(MV)/Y"] - data["P"]
data[["EQ9"]] = data["V/Y"] - data["P/M"]
data[["EQ10"]] = data["V"] - data["(PY)/M"]
data[["EQ11"]] = data["V/P"] - data["Y/M"]
data[["EQ12"]] = data["V/(PY)"] - data["1/M"]
data[["EQ13"]] = data["1/(PY)"] - data["1/(MV)"]
data[["EQ14"]] = data["1/P"] - data["Y/(MV)"]
data[["EQ15"]] = data["1"] - data["(PY)/(MV)"]
data[["EQ16"]] = data["1/Y"] - data["P/(MV)"]
```
