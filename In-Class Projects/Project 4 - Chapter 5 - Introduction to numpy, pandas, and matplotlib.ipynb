# Create random data for a dataframe

```python
macro_dict = {"GDP":{},
            "Money":{},
#            "Real GDP":{},
            "Price Level":{}}
```

```python
for key in macro_dict:
    for i in range (1990, 2018):
        macro_dict["GDP"][i] = np.random.random() * 20000
        macro_dict[key][i] = np.random.random() * 10000
```

```python
macro_dict
```

```python
macro_df = pd.DataFrame(macro_dict)
macro_df
```

```python
macro_df["Velocity"] = macro_df["GDP"].div(macro_df["Money"])
macro_df["Real GDP"] = macro_df["GDP"].div(macro_df["Price Level"])
macro_df
```

## You can call a subset of rows by passing a list of keys

```python
macro_df[["Real GDP" , "Money"]]
```

```python
macro_df.loc[1999]
```

```python
macro_df.loc[1999:2005, ["Real GDP" , "Money"]]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
macro_df.plot.line()
```

```python
for key in macro_df:
    fig, ax = plt.subplots(figsize = (12,8))
    macro_df[key].plot.line(ax = ax)
    macro_df[key].plot.line(ls = "" , marker = ".", ax = ax)
    plt.title(key, fontsize = 20)
    plt.show()
    plt.close()
```

```python
for key in macro_df:
    fig, ax = plt.subplots(figsize = (12, 8))
    macro_df[key].plot.line(color= "C0", ax = ax)
    macro_df[key].plot.line(ls= "", marker = ".", color = "C1", ax = ax)
    for ix, val in zip(macro_df.index, macro_df[key]):
        ax.text(ix, val, str(round(val,2)), color = "C4")
    plt.title(key, fontsize = 22)
    plt.show()
    plt.close()
```
